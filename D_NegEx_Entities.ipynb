## Daniel's NegEx Function (Entity Based Implementation)

```python
#Use entity patterns to implement negation (future update: and pseudo-negation)
#Negation acts forward 5 tokens

#Future update?: Reverse negation terms with reverse scope

#Currently does NOT care about end of sentences, does NOT deal with context, does NOT include pseduo or
#reverse negations

spacy.tokens.Span.set_extension( "negation", default = "affirmed", force=True)

def negEx (doc, range = 5):
    for sentence in doc.sents:
        for switchEnt in sentence.ents:
            #if pseudo
            #short circuit, don't check for negation entities following (range)
            #above pseduocode might be unnecessary
            if switchEnt.label_ == "NEG":
                for potentialNegEnt in doc[switchEnt.end:switchEnt.end + range].ents:
                    potentialNegEnt._.negation = "negated"               
    return doc

nlp.remove_pipe("NegEx")
nlp.add_pipe(negEx, "NegEx", last = True)
#the following makes sure that entities are treated as a single token. Fixes issues with negation span
nlp.remove_pipe("merge_entities")
nlp.add_pipe(nlp.create_pipe("merge_entities"))
```
